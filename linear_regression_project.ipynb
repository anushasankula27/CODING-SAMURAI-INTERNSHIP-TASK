```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data_dict = {
    "Advertising_Spend": [100, 150, 200, 250, 300, 350, 400, 450, 500],
    "Sales": [10, 14, 18, 22, 26, 30, 34, 38, 42]
}
df = pd.DataFrame(data_dict)
df.to_csv("advertising_sales.csv", index=False)
print("CSV file created successfully!\n")
data = pd.read_csv("advertising_sales.csv")
print("Dataset Preview:")
print(data.head(), "\n")
X = data[['Advertising_Spend']]
y = data['Sales']
model = LinearRegression()
model.fit(X, y)
print("Model trained successfully!\n")
predicted_sales = model.predict(X)
m = model.coef_[0]
c = model.intercept_
print("Linear Regression Equation:")
print(f"Sales = {m:.2f} * Advertising_Spend + {c:.2f}\n")
new_spend = [[600]]
new_prediction = model.predict(new_spend)
print(f"Predicted Sales for Advertising Spend = 600 : {new_prediction[0]:.2f}\n")
plt.scatter(X, y)
plt.plot(X, predicted_sales)
plt.xlabel("Advertising Spend")
plt.ylabel("Sales")
plt.title("Advertising Spend vs Sales (Linear Regression)")
plt.show()
```
